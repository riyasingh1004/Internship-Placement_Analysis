## k Nearest Neighbors Classification Model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
import warnings
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore', category=DataConversionWarning)
```

```python
k_range = range(3,10)
knn_scores = {}
knn_scores_list = []
```

```python
for k in k_range:
    neighbor = KNeighborsClassifier(n_neighbors=k)
    neighbor.fit(X_train, y_train)
    y_pred = neighbor.predict(X_test)
    knn_y_pred = pd.DataFrame(y_pred,columns = ['status_Placed'])
    knn_scores[k] = metrics.accuracy_score(y_test, knn_y_pred)
    knn_scores_list.append(metrics.accuracy_score(y_test, knn_y_pred))
```

#### In KNN, K is the number of nearest neighbors. The number of neighbors is the core deciding factor.

```python
knn_y_pred.sample(5)
```

```python
plt.plot(k_range, knn_scores_list)
plt.xlabel('Value of k for knn')
plt.ylabel('Testing scores')
```

#### Here the plot shows value of k vs accuracy score for different values of k.  We can observe that we are getting highest accuracy at k=5,7,9  i.e 0.909 whic can be considered valid as when number of classes(here 'Placed' and 'Not Placed') are even,k is generally odd.

```python
for key,value in knn_scores.items():
    print(key,':',value)
print('max:',max(knn_scores_list))
```

## Decision Tree Classifier model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc_range = range(3,10)
dtc_scores = {}
dtc_scores_list = []
```

```python
for k in dtc_range:
    tree = DecisionTreeClassifier()
    tree.fit(X_train, y_train)
    dtc_y_pred = tree.predict(X_test)
    
    dtc_y_pred = pd.DataFrame(dtc_y_pred,columns = ['status_Placed'])
    dtc_scores[k] = metrics.accuracy_score(y_test, dtc_y_pred)
    dtc_scores_list.append(metrics.accuracy_score(y_test, dtc_y_pred))
```

#### A decision tree is a flowchart-like tree structure where an internal node represents feature(or attribute), the branch represents a decision rule, and each leaf node represents the outcome. The topmost node in a decision tree is known as the root node. It learns to partition on the basis of the attribute value. It partitions the tree in recursively manner call recursive partitioning.

```python
plt.plot(dtc_range, dtc_scores_list)
plt.xlabel('Value of k for knn')
plt.ylabel('Testing scores')
```
